```python
# f(x) = e**x - log(x)
# has only one minimum point (global minimum)


# df/dx(x0) -> gradient
# initial point x0
# x1 = x0 - a * df/dx(x0)

# a>0, iteration times is also a parameters


def f_example_1(x):
    return np.exp(x) - np.log(x)


def dfdx_example(x):
    return np.exp(x) - 1/x
```

```python

plot_f([0.001, 2.5], [-0.3, 13], f_example_1, 0.0)
```

```python
def gradient_descent(dfdx, x, learning_rate=0.1, num_iterations=100):
    for _ in range(num_iterations):
        x = x - learning_rate * dfdx(x)
    
    return x
```

```python
num_iterations = 25; learning_rate = 0.1; x_initial = 1.6
print("Gradient descent result: x_min =", gradient_descent(dfdx_example, x_initial, learning_rate, num_iterations)) 
```
